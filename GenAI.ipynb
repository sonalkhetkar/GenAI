```python
import sqlite3

conn = sqlite3.connect("school.db")
cursor = conn.cursor()

sql_files = [
    "school_db_bankdetails.sql",
    "school_db_classes.sql",
    "school_db_marks.sql",
    "school_db_parents.sql",
    "school_db_scholarships.sql",
    "school_db_students.sql",
    "school_db_subjects.sql"
]

base_path = "/content/sql_extract/Dump20250716/Dump20250716"

for file in sql_files:
    path = os.path.join(base_path, file)
    print(f" Importing: {file}")
    with open(path, "r", encoding="utf-8", errors="ignore") as f:
        sql_script = f.read()
        cursor.executescript(sql_script)

conn.commit()

cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = cursor.fetchall()
print("Tables in school.db:", [t[0] for t in tables])

conn.close()
```
